## Sentiment Analyzis using Transformers Models

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier (
    [
        "I've been waiting for a HuggingFace course my whole life.",
        "I hate this so much!"
    ]
)
```

## Text generation model like ChatGPT

```python
# Text generation model like ChatGPT
generator = pipeline("text-generation")
generator(
    "I love the Jurassic Park movies! I think the original Jurassic Park is a masterpiece. ",
    max_length=50,
    num_return_sequences=1,
)
```

## QnA Model

```python
qa = pipeline("question-answering")
qa (
    question="Who is the protagonist of Jurassic Park?",
    context = "Dr. Allel Grant saves the crew using his 3D printed Velociraptor SoundBox...",
)
```

## Language Translator

```python
# Language Translator
translator = pipeline("translation_en_to_de")
translator("I love Samriddhi!")
```

## NER Models

```python
ner = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english")
ner("i am Shlok and i live in Lalitpur")
```
